# (1) 주제 선정 이유


오늘날 자본주의를, 그리고 우리가 사는 세상을 빠르게 발전시킨 요소의 밑바탕에는 항상 시장이 있었다. 원시적 시장형태인 개인 대 개인의 
물물교환부터, 현재의 다수의 공급자와 다수의 수요자가 존재하는 시장까지 발전하게 되었다. 또한 공급자는 본인이 가진 자산만을 이용하는 것이 아니라, 투자자들의 자산을 이용해 본인의 기업을 운영해 나가는 사회형태까지 발전하게 되었다. 투자자들은 본인들이 투자한 금액에 대한 금전적인 책임만 따를 뿐, 그 외의 책임이 따르지 않는, 책임을 최소화하고 여러가지 위험성을 내포한 채, 발전의 극대화에 최적화된 사회형태가 되었다.<br> 
( 이러한 형태의 시장 발전은 어떻게 보면 더 나은 것을 추구하는 인간의 욕심에 대응해 만들어진 불가결한 형태였는지도 모르겠다.)

본래 일반적으로, 투자를 하는 대부분의 사람들은 투자를 함으로써 돌아오는 이익을 얻기 위함이 목적이다. ( 투자를 하면서 본인이 이 회사에 투자함으로써 이 회사가 사회에 공헌할 이익에 내가 기여한다 생각하며 투자할 투자자들이 몇이나 될까? ) 즉, 투자자 개개인들은 돈을 벌기 위함이 그 목적이라고 할 수 있다. 돈이라고 하는 물질은 세상의 필요악과 같은 존재로, 사람들의 옆에서 웃게 하기도 하고 울게 하기도 하며 비참하게 만들기도 하고 감동을 주기도 하는 수단이다. 돈에 대한 욕망을 근거로 해서 투자시장은 커져 갔다고 해도 과언이 아닐 것이다. 본인이 노동을 하는 것이 아니라, 본인이 가진 자금을 통해 누군가를 노동시키고 거기서 얻어지는 수익의 대가를 지불받는, 투자자로서는 상당히 매력적인 수익구조가 아닐 수 없다. 수많은 금융회사들, 그로부터 파생되는 금융상품들은 이런 인간의 욕망을 근거로 성장해왔다고 볼 수 있다. 

한편, 기술이 날이 갈수록 발전하고 이에 따라 시장들이 다분화되면서 사람들은 특정 시장들에 대한 전문적인 지식을 얻기도 어려워 질 뿐더러, 시장과 시장 사이의 연관성, 사회의 전반적인 흐름까지 고려하여 투자하여야 하는 상황에 이르게 되었다. 기술이 발전할수록 투자할 수 있는 부분은 많아지지만 어디에 투자해야할지는 개인으로서 막연한 상태에 이른 것이다. 이 과정에서 그런 막연함을 막아주기 위해 투자회사들이 등장하게 되고, 금융상품들이 등장하게 된다. 투자회사들은 각각 나름의 경제분석 레포트나 기업분석 레포트를 내는 등 각 회사들의 관점을 더한 레포트를 내며 다수의 투자할 돈은 있지만 어디에 투자해야 할지 모르는 사람들에게 매력적인 요인으로 작용하게 된다. 

오늘날 투자의 근거로는 두가지를 들 수가 있다. 첫째, 기업의 재무제표를 분석함으로써 기업이 기업활동을 수행하는 과정 중에서 발생한 돈의 흐름들에 대해 살펴보는 것이다. 이는 이미 발생한 사건에 대한 정보로 과거의 정보라고 할 수 있지만 이런 데이터를 바탕으로 기업이 가지고 있는 시장에서의 위치, 기업의  성장 정도 등을 예측해 볼 수 있을 것이다. 둘째, 더욱 가까운 현재에 발생하고 있는 사회적 사건들을 통해 투자의 근거를 얻을 수 있다. 예를 들어, A 기업의 신기술 개발 성공으로 개발비용은 감소하고 더욱 효율적으로 재화를 생산할 수 있게 되었다는 기사가 떴다고 가정해보자. 그렇다면 다수의 투자자들은 A 기업에 평소보다 많은 관심을 기울이며 투자를 하게 될 것이다. 

이번 프로젝트에서는 몇년간의 기업들의 활동을 돈의 흐름으로 나타낸 재무제표를 분석해 봄으로써 재무분석지표들의 값을 도출해 낼 것이다. 도출해 낸 지표들의 값들은 여러 방면으로 의미있게 사용될 수 있을 것이다. **첫째로는, 투자자들에게 재무제표(가공되지 않은 데이터)로부터 기업의 건전성, 투자성 등(가공된 데이터)의 데이터를 제공하여 보다 쉽게 투자 판단에 도움이 되고자 하는 바이다.** 이런 프로그램이 무료로 배포된다면 경제적으로도 긍정적으로 작용할 수 있을 것으로 기대된다.`#(1)` 이러한 논리적, 수치적 자료가 없어 투자를 망설이던 투자자들에게는 정보를 획득할 수 있는 프로그램으로 작용할 것이며, 가공된 정보가 수월하게 제공됨으로써 투자를 장려할 수 있을 것이다. 이는 자연스럽게 경제 선순환에 이바지하는 프로그램이 될 수 있을 것이라 생각한다. **두번째로는, 투자부분의 목적이 아니라, 관리 및 감독 측면에서도 활용할 수 있을 것이다.`**예를 들어, 최근에 있었던 삼성바이오로직스 분식회계 사건을 예로 들 수 있다. (주)삼성바이오로직스 사가 상장을 하면서 영업이익이 계속해서 적자가 나던 회사구조에서 1년 사이에 2조 이상의 이익을 남기게 된다. 어떤 회사든 1년 사이에 적자에서 2조 이상의 이익이 난 것은 의심을 받을 만한 사항이다. 재무제표의 자료들로부터 데이터들이 가공되지 않았다면 분식회계 사건의 내막을 조사할 여건조차 되지 않았을 것이다. 관리 및 감독 부분에 있어, 이러한 특이점들에 있어 쉽게 경고를 보내줄 수 있는 프로그램이 있다면, 이러한 측면에서 또한 조금은 수월한 조사가 가능할 것이다.     


`#(1)` GDP는 크게 소비지출, 투자지출, 정부지출, 순수출의 합으로 계산된다. 즉, 투자지출이 증가한다면 GDP의 증가로 이어진다. GDP의 증가는 주로 경제성장을 의미한다.  

# (2) 가설 정의


기업들의 재무제표에서 얻어진 데이터를 기반으로 해당 기업을 분석한다. 기업의 재무제표에는 가공되지 않은 데이터들이 존재한다. 재무제표에서 특정 데이터들을 뽑아내 유의미한 값으로 계산해 내어 투자자들이 해당 기업에 대한 투자 여부를 결정할 때에 결정적인 근거들을 만들어 주고자 한다. 재무제표를 분석하여 기업의 성장성, 건전성 등을 파악함에 따라 현재 시점부터 벌어질 기업의 기업활동들을 조금이나마 예측해 볼 수 있을 것이다. 

또한 삼성바이오로직스 분식회계 사건이 왜 문제가 되었는지에 대해 그 근거를 재무제표로부터 찾아나가 보려 한다. 이런 점들을 재무제표를 통해 분석해 볼 수 있다는 것은, 즉 재무제표를 통해 기업의 기업활동이 올바르게 행해지고 있는지 또한 감시할 수 있다는 것이다. 회계처리의 일반적인 특성들을 도출해내고 그 특성들에 어긋나는 회계자료들에 대해 의심을 품는 근거 자료를 제시할 수 있는 프로그램을 만든다.

재무제표를 분석하는 데 있어, 중요한 근거가 되는 자료들은 `재무제표 비율분석`자료들이다. 기업의 재무제표에서 얻은 재무비율들을 통해 기업의 재무건정성과 성과를 분석할 수 있다. 재무비율 분석 중 주요 재무비율로 여겨지는 요소들에 대해서 재무제표로부터 데이터를 가공할 것이다.

# (3) 인터넷을 통한 데이터 획득

# (4) 분석을 위한 데이터의 가공  

나는 엑셀파일을 불러와 데이터를 가공하는 과정을 **`1차 가공과정`**과 **`2차 가공과정`**으로 나누었다. <br>
여기서 **`1차 가공과정`**이란, `엑셀파일을 파이썬 환경으로 불러와 가공에 알맞은 데이터 프레임을 생성하는 과정`으로 정의하였다. <br>
**`2차 가공과정`**은, `1차 가공된 데이터프레임의 데이터들을 가지고 원하는 값들을 추출해 내 최종적으로 원하는 데이터를 얻어내는 과정`을 말한다.
<br><br>

### 1)   1차 가공과정

**1차 가공과정**<br><br>
(1) 사업보고서 파일에서 엑셀 파일의 재무제표를 다운로드한다. <br>
(2) 엑셀 파일을 openpyxl module과 pandas moudle을 활용해 DataFrame으로 전환한다.  <br>
(3) DataFrame의 2차 가공을 수월하기 하기 위해 DataFrame의 index와 header 데이터를 의미있는 값으로 수정한다.<br>

1차 가공과정에서는, 쓸모없는 값들을 데이터 프레임 안에 존재하지 못하게 하는 작업을 수행한다. 재무제표에서 의미있는 값을 가지기 시작하기 전의 행은 단위를 나타내는 부분으로 항상 동일하다. 특정 행에서 단위를 나타내는 부분이 존재하게 되고 그 부분까지 데이터 프레임에서 제거하여 새로운 데이터 프레임을 구성하게 한다. 그렇게 의미없는 값을 제거한 뒤에는 index와 header부분을 의미있는 값으로 다시 설정해 주어야 한다. index로는 각각의 값들이 무슨 값을 의미하는지 데이터의 종류에 대해서 명시해 준다. header에 대해서는 그 데이터를 가지는 시기 즉, 년도가 언제인지에 대해 명시할 수 있도록 한다. 재무제표 상에서는 '기'라는 용어를 사용하는데, 이는 상장된 뒤부터 카운팅을 한 값이다.<br> 

1차 가공과정에서 사용한 방법을 간단하게 이야기하자면, 우선은 인덱스의 길이를 세고, 인덱스로 반복문을 돌려 해당 인덱스에서 가지는 값이 의미가 있어지기 전까지 즉, 마지막으로 의미가 없는 곳의 그 위치를 위치를 찾는다. 그리고 해당 인덱스까지는 drop 하여 데이터프레임을 수정한다. 수정된 인덱스는 첫번째 인덱스에서 의미있는 값을 가지는 인덱스가 될텐데, 해당 인덱스에서 가지는 값은 이 후 헤더 역할을 할, 기수 정보가 될 것이다. 따라서 기수 정보를 헤더로 올려준다. 그 다음에는 숫자로 구성된 인덱스를 두 번째 열에 있을 의미있는 값의 종류들을 당겨와, 인덱스로 설정해주는 작업을 수행한다. 

한편, 모든 기업의 사업보고서에 관해 적용될 수 있도록 프로그램을 구현하고자 하였고 이를 보이기 위해 결론에서는 대표적으로 3개의 기업들에 대해 재무제표를 분석하는 과정을 보여준다. <br><br>

```python
import openpyxl 
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
import matplotlib.font_manager as fm 
from makeplt import *
####    한글폰트설정    ####
font_fname = "C:\\Users\\chon0\\Anaconda3\\envs\\develroom\\lib\\site-packages\\matplotlib\\mpl-data\\fonts\\ttf\\Lotte_Dream_bold.ttf"
fontprop = fm.FontProperties(fname=font_fname)
####    ###########    #### 


#본격적인 데이터 추출을 위한 엑셀 리딩과 가공이 쉽도록 데이터프레임을 형성 
# 그냥 불러다 쓰면 된다. 이미 가공이 쉽게 CIS 와 FS 정리 
class total_FS :
    del_rowFS = 0
    del_rowCIS = 0 
    
    # self.year / self.xlsx / self.df_CIS / self.df_FS / 
    # __init__  /  sefFS  / setCIS  /  xlsxtoFS  / xlsxtoCIS  /  is_proc1st_CISver  /  is_proc1st_FSver  /  proc1st_FSver / proc1st_CISver

    # when initialiazing , can get df_CIS & df_FS
    def __init__(self,year,xlsx):
        self.year = year
        self.xlsx = xlsx
        self.df_CIS = total_FS.xlsxtoCIS(self)
        self.df_FS = total_FS.xlsxtoFS(self)
        
        if total_FS.is_proc1st_CISver(self):
            total_FS.proc1st_CISver(self)
            total_FS.proc2nd_CISver(self)
            total_FS.change_index_CIS(self)
        else : 
            total_FS.change_index_CIS(self)

        if total_FS.is_proc1st_FSver(self):
            total_FS.proc1st_FSver(self)
            total_FS.proc2nd_FSver(self)
            total_FS.change_index_FS(self)
        else : 
            total_FS.change_index_FS(self)
        
    def setFS(self,FS_xlsx):
        self.FS = FS_xlsx

    def setCIS(self,CIS_xlsx):
        self.CIS = CIS_xlsx

    def xlsxtoFS (self):        #make df_FS from xlsx 

        sheetList = [] 

        wb = openpyxl.load_workbook(self.xlsx)

        for i in wb.get_sheet_names():
            sheetList.append(i)

        xlsx = pd.ExcelFile(self.xlsx)

        if "연결 재무상태표" in sheetList :
            for i in range(len(sheetList)):
                if (sheetList[i] == "연결 재무상태표"):
                    df_FS = pd.read_excel(xlsx,sheetList[i])
            return df_FS
        else :
            for i in range(len(sheetList)):
                if (sheetList[i] == "재무상태표"):
                    df_FS = pd.read_excel(xlsx,sheetList[i])
            return df_FS

    def xlsxtoCIS(self):        #make df_CIS from xlsx 
        sheetList = [] 

        wb = openpyxl.load_workbook(self.xlsx) ##

        for i in wb.get_sheet_names():
            sheetList.append(i)

        xlsx = pd.ExcelFile(self.xlsx) ## 

        if "연결 손익계산서" in sheetList :
            for i in range(len(sheetList)):
                if (sheetList[i] == "연결 손익계산서"):
                    df_CIS = pd.read_excel(xlsx,sheetList[i])
            return df_CIS
        elif "연결 포괄손익계산서" in sheetList :
            for i in range(len(sheetList)):
                if (sheetList[i] == "연결 포괄손익계산서"):
                    df_CIS = pd.read_excel(xlsx,sheetList[i])
            return df_CIS
        elif "손익계산서" in sheetList :
            for i in range(len(sheetList)):
                if (sheetList[i] == "손익계산서"):
                    df_CIS = pd.read_excel(xlsx,sheetList[i])
            return df_CIS
        else :
            for i in range(len(sheetList)):
                if (sheetList[i] == "포괄손익계산서"):
                    df_CIS = pd.read_excel(xlsx,sheetList[i])
            return df_CIS


    def is_proc1st_CISver(self):        #checking the need to process 
        k=0
        if ((self.df_CIS.columns[0]!='과                        목') & (self.df_CIS.columns[0]!='   ')):
            for j in range(len(self.df_CIS.index.values)):
                if((self.df_CIS.iloc[[j],[0]].values[0][0]=="(단위 : 원)")|(self.df_FS.iloc[[j],[0]].values[0][0]=="(단위 : 백만원)")):
                    k = j        
        if(k!=0):
                #df = df.drop(df.index[list(range(k+1))])
            return True 
        else:
            return False

    def is_proc1st_FSver(self):         #checking the need to process 
        k=0
        if ((self.df_FS.columns[0]!='과                        목') & (self.df_FS.columns[0]!='   ')):
            for j in range(len(self.df_FS.index.values)):
                #print(i.iloc[[j],[0]].values[0][0])
                if((self.df_FS.iloc[[j],[0]].values[0][0]=="(단위 : 원)")|(self.df_FS.iloc[[j],[0]].values[0][0]=="(단위 : 백만원)")):
                    k = j        
        if(k!=0):
                #df = df.drop(df.index[list(range(k+1))])
            return True 
        else:
            return False

    def proc1st_FSver(self):
        k=0
        if ((self.df_FS.columns[0]!='과                        목') & (self.df_FS.columns[0]!='   ')):
            for j in range(len(self.df_FS.index.values)):
                #print(i.iloc[[j],[0]].values[0][0])
                if((self.df_FS.iloc[[j],[0]].values[0][0]=="(단위 : 원)")|(self.df_FS.iloc[[j],[0]].values[0][0]=="(단위 : 백만원)")):
                    k = j        
                    total_FS.del_rowFS = j
            if(k!=0):
                self.df_FS = self.df_FS.drop(self.df_FS.index[list(range(k+1))])
        return self.df_FS


    def proc1st_CISver(self):
        k=0
        if ((self.df_CIS.columns[0]!='과                        목') & (self.df_CIS.columns[0]!='   ')):
            for j in range(len(self.df_CIS.index.values)):
                #print(i.iloc[[j],[0]].values[0][0])
                if((self.df_CIS.iloc[[j],[0]].values[0][0]=="(단위 : 원)")|(self.df_CIS.iloc[[j],[0]].values[0][0]=="(단위 : 백만원)")):
                    k = j        
                    total_FS.del_rowCIS = j
            if(k!=0):
                self.df_CIS = self.df_CIS.drop(self.df_CIS.index[list(range(k+1))])
        return self.df_CIS
        

    def proc2nd_FSver(self):    #change header & change index as category    #it must be implement after 1st process 
        sign = self.df_FS.index[0] #4  #index의 0번째에 있는 값이 무슨 값인지 반환 
        re_header =[]
        for i in range(len(self.df_FS.loc[[sign]].values[0])):
            re_header.append(self.df_FS.loc[[sign]].values[0][i])

        self.df_FS.columns = re_header 
        self.df_FS = self.df_FS.drop(self.df_FS.index[[0]])
        self.df_FS = self.df_FS.rename(index={})
        return self.df_FS
        

    def proc2nd_CISver(self):    #change header & change index as category #it must be implement after 1st process
        sign = self.df_CIS.index[0] #4  #index의 0번째에 있는 값이 무슨 값인지 반환     
        re_header =[]
        for i in range(len(self.df_CIS.loc[[sign]].values[0])):
            re_header.append(self.df_CIS.loc[[sign]].values[0][i])

        self.df_CIS.columns = re_header 
        self.df_CIS = self.df_CIS.drop(self.df_CIS.index[[0]])
        return self.df_CIS

    def change_index_CIS(self):
        #change the index as category 
        # df_2017FS.rename(index={df_2017FS.index[0]:df_2017FS.iloc[[1],[0]].values[0][0]})

        for i in range(len(self.df_CIS.index)):
            self.df_CIS = self.df_CIS.rename(index = {self.df_CIS.index[i]:self.df_CIS.iloc[[i],[0]].values[0][0]})
        self.df_CIS = self.df_CIS.drop(self.df_CIS.columns[0],axis=1)
        return self.df_CIS

    def change_index_FS(self):
        #change the index as category 
        # df_2017FS.rename(index={df_2017FS.index[0]:df_2017FS.iloc[[1],[0]].values[0][0]})

        for i in range(len(self.df_FS.index)):
            self.df_FS = self.df_FS.rename(index = {self.df_FS.index[i]:self.df_FS.iloc[[i],[0]].values[0][0]})
        self.df_FS = self.df_FS.drop(self.df_FS.columns[0],axis=1)
        return self.df_FS
```

```python
class Finance_ratio_from_audit (total_FS):         ##사업보고서가 아닌 감사보고서로부터 따온 재무비율 분석을 위한 클래스 #사실상 삼바 계산위해,,

    def __init__(self,year,xlsx):
        super().__init__(year,xlsx)
        self.flow_rate_ = Finance_ratio_from_audit.flow_rate(self)
        self.cur_rate_ = Finance_ratio_from_audit.current_rate(self)


    def flow_rate (self):      #유동성비율  #유동비율 = 유동자산/유동부채 *100 (%)
        liq_asset_index = []
        liq_liab_index =[]
        # '유동자산' 과 '유동부채'라는 단어를 포함하는 요소가 많음
        # 하지만 재무제표 특성 상 '유동자산'과 '유동부채' 항목은 제일 첫번째에 있으므로, 첫번째 인덱스 항목이용
        for i in range(len(self.df_FS.index)):           
            if(set('유동자산').issubset(set(self.df_FS.index[i]))):
                liq_asset_index.append(i)
            if(set('유동부채').issubset(set(self.df_FS.index[i]))):
                liq_liab_index.append(i)
        #dictionaty 생성 후, 년도별 요소 추가
        flow_rate={}
        for i in range(2):  # i = 1, 3 
            liq_asset = self.df_FS.iloc[[liq_asset_index[0]],[i*2+1+1]].values[0][0]
            liq_liab = self.df_FS.iloc[[liq_liab_index[0]],[i*2+1+1]].values[0][0]
            flow_= round((liq_asset / liq_liab) *100,3)
            flow_rate[str(self.year-i)+"_flow_rate"] = flow_
        return flow_rate

    def current_rate (self):    #유동성 비율   # 당좌비율 = (유동자산 - 재고자산)/유동부채 * 100 (%) 
        liq_asset_index = []
        liq_liab_index =[]
        liq_inven_index =[]
        for i in range(len(self.df_FS.index)):
            if(set('유동자산').issubset(set(self.df_FS.index[i]))):
                liq_asset_index.append(i)
            if(set('유동부채').issubset(set(self.df_FS.index[i]))):
                liq_liab_index.append(i)
            if(set('재고자산').issubset(set(self.df_FS.index[i]))):
                liq_inven_index.append(i)                                                   
        current_rate={}
        for i in range(2):
            liq_asset = self.df_FS.iloc[[liq_asset_index[0]],[i*2+1+1]].values[0][0]                  
            liq_liab = self.df_FS.iloc[[liq_liab_index[0]],[i*2+1+1]].values[0][0]                    
            liq_inven = self.df_FS.iloc[[liq_inven_index[0]],[i*2+1]].values[0][0]                   
            current_ = round((liq_asset - liq_inven) / liq_liab *100 ,3)
            current_rate[str(self.year-i)+"_current_rate"] = current_
        return current_rate
        

    def leverage_rate (self):       #부채비율       #레버리지비율 = 총부채 / 총 자본 *100 (%)
        total_debt_index =[]
        total_capi_index =[]
        for i in range(len(self.df_FS.index)):
            if(set('부채총계').issubset(set(self.df_FS.index[i]))):
                total_debt_index.append(i)
            if(set('자본총계').issubset(set(self.df_FS.index[i]))):
                total_capi_index.append(i)
        leverage_rate ={}
        for i in range(2):
            total_debt = self.df_FS.iloc[[total_debt_index[0]],[i*2+1+1]].values[0][0]
            total_capi = self.df_FS.iloc[[total_capi_index[0]],[i*2+1+1]].values[0][0]
            lever_ = round(total_debt/total_capi *100,3)
            leverage_rate[str(self.year -i)+"_leverage_rate"] = lever_ 
        return leverage_rate

    def sales_incre_rate (self):        #성장성 비율    #매출액증가율 = (당기매출액-전기매출액)/전기매출액 * 100 
        sales_incre_rate = {}
        sales ={}
        sales_index =[]
        for i in range(len(self.df_CIS.index)):
            if(set('매출액').issubset(set(self.df_CIS.index[i]))):
                sales_index.append(i) 
            elif(set('영업수익').issubset(set(self.df_CIS.index[i]))):
                sales_index.append(i) 
        for i in range(2):
            sales[str(self.year -i)] = self.df_CIS.iloc[[sales_index[0]],[i*2+1+1]].values[0][0] ##year , year-1 , year-2 생성 
        for i in range(len(sales)-1):
            sales_incre_rate[str(self.year -i)+"_sales_incre_rate"] = round(((sales[str(self.year-i)]-sales[str(self.year-i-1)])/sales[str(self.year-i-1)] )*100 ,3)
        return sales_incre_rate

    def net_profit_rate (self):         #성장성 비율    #순이익증가율  = (당기순이익 - 전기순이익) / 전기순이익 *100 
        net_profit_rate ={}
        net_pro={}
        net_pro_index =[]
        for i in range(len(self.df_CIS.index)):
            if(set('당기순이익').issubset(set(self.df_CIS.index[i]))):
                net_pro_index.append(i)  
        for i in range(2):
            net_pro[str(self.year - i)] = self.df_CIS.iloc[[net_pro_index[0]],[i*2+1+1]].values[0][0]
        for i in range(len(net_pro)-1):
            net_profit_rate[str(self.year -i)+"_net_profit_rate"] = round(((net_pro[str(self.year-i)]-net_pro[str(self.year-i-1)])/net_pro[str(self.year-i-1)]) *100 ,3)
        return net_profit_rate

    def ROA (self):     #수익성 비율    #총자산 순이익률(Return On Assets , ROA)  = 당기순이익/총자산 *100 
        net_pro={}
        total_asset ={}
        ROA ={}
        total_asset_index =[]
        net_pro_index=[]
        for i in range(len(self.df_FS.index)):
            if(set('자산총계').issubset(set(self.df_FS.index[i]))):
                total_asset_index.append(i)
        for i in range(len(self.df_CIS.index)):
            if(set('당기순이익').issubset(set(self.df_CIS.index[i]))):
                net_pro_index.append(i)  
        for i in range(2):
            total_asset[str(self.year -i)] = self.df_FS.iloc[[total_asset_index[0]],[i*2+1+1]].values[0][0]
        for i in range(2):
            net_pro[str(self.year - i)] = self.df_CIS.iloc[[net_pro_index[0]],[i*2+1+1]].values[0][0]
        for i in range(len(net_pro)):
            ROA[str(self.year - i)+"_ROA"] = round((net_pro[str(self.year-i)]/total_asset[str(self.year -i)])*100 ,3)
        return ROA 

    def operating_profit_rate (self):       #수익성 비율       #매출액 영업이익률 = 영업이익/매출액 *100 
        sales ={}
        oper_pro ={}
        oper_pro_rate ={}
        sales_index =[]
        oper_pro_index =[]
        for i in range(len(self.df_CIS.index)):
            if(set('매출액').issubset(set(self.df_CIS.index[i]))):
                sales_index.append(i) 
            elif(set('영업수익').issubset(set(self.df_CIS.index[i]))):
                sales_index.append(i)
        for i in range(2):
            sales[str(self.year -i)] = self.df_CIS.iloc[[sales_index[0]],[i*2+1+1]].values[0][0] ##year , year-1 , year-2 생성

        for i in range(len(self.df_CIS.index)):
            if(set('영업이익').issubset(set(self.df_CIS.index[i]))):
                oper_pro_index.append(i)
            elif(set('영업손실').issubset(set(self.df_CIS.index[i]))):
                oper_pro_index.append(i)
        for i in range(2):
            oper_pro[str(self.year -i)] = self.df_CIS.iloc[[oper_pro_index[0]],[i*2+1+1]].values[0][0]

        for i in range(len(oper_pro)):
            oper_pro_rate[str(self.year -i)+"_oper_pro_rate"] = round((oper_pro[str(self.year - i)]/sales[str(self.year -i)])*100 ,3)

        return oper_pro_rate

    def ATR (self):         #활동성 비율    #총자산회전율(Asset Turnover Ration, ATR) = 총매출액/총자산  
        sales ={}
        total_asset ={}
        ATR = {}
        sales_index = []
        total_asset_index = []
        for i in range(len(self.df_CIS.index)):
            if(set('매출액').issubset(set(self.df_CIS.index[i]))):
                sales_index.append(i)  
            elif(set('영업수익').issubset(set(self.df_CIS.index[i]))):
                sales_index.append(i)
        for i in range(2):
            sales[str(self.year -i)] = self.df_CIS.iloc[[sales_index[0]],[i*2+1+1]].values[0][0]

        for i in range(len(self.df_FS.index)):
            if(set('자산총계').issubset(set(self.df_FS.index[i]))):
                total_asset_index.append(i) 
        for i in range(2):
            total_asset[str(self.year -i)] = self.df_FS.iloc[[total_asset_index[0]],[i*2+1+1]].values[0][0]

        for i in range(len(total_asset)):
            ATR[str(self.year - i)+"_ATR"] = round(sales[str(self.year - i)]/total_asset[str(self.year - i)],3)

        return ATR
    
    def FAT (self) :    #활동성 비율    #고정자산회전율(Fixed Asset Turnover, FAT) = 매출액/비유동자산 
        sales ={}
        non_cur_asset ={}
        FAT ={}
        sales_index = []
        noncur_ass_index=[]
        for i in range(len(self.df_CIS.index)):
            if(set('매출액').issubset(set(self.df_CIS.index[i]))):
                sales_index.append(i)
            elif(set('영업수익').issubset(set(self.df_CIS.index[i]))):
                sales_index.append(i)
        for i in range(len(self.df_FS.index)):
            if(set('비유동자산').issubset(set(self.df_FS.index[i]))):
                noncur_ass_index.append(i) 
        for i in range(2):
            sales[str(self.year -i)] = self.df_CIS.iloc[[sales_index[0]],[i*2+1+1]].values[0][0]
        for i in range(2):
            non_cur_asset[str(self.year -i)] = self.df_FS.iloc[[noncur_ass_index[0]],[i*2+1+1]].values[0][0]
        for i in range(len(non_cur_asset)):
            FAT[str(self.year -i)+"_FAT"] = round(sales[str(self.year -i)]/non_cur_asset[str(self.year -i)],3)

        return FAT
        
    def RTR (self) :    #활동성비율     #매출채권 회전율(회) (Receivables Turnover Ratio) = 매출액/매출채권 
        sales ={}
        acc_recei = {}
        sales_index = []
        acc_recei_index =[]
        RTR ={}
        for i in range(len(self.df_CIS.index)):
            if(set('매출액').issubset(set(self.df_CIS.index[i]))):
                sales_index.append(i)
            elif(set('영업수익').issubset(set(self.df_CIS.index[i]))):
                sales_index.append(i)
        for i in range(len(self.df_FS.index)):
            if(set('매출채권').issubset(set(self.df_FS.index[i]))):
                acc_recei_index.append(i)

        for i in range(2):
            sales[str(self.year -i)] = self.df_CIS.iloc[[sales_index[0]],[i*2+1+1]].values[0][0]
        for i in range(2):
            acc_recei[str(self.year - i)] = self.df_FS.iloc[[acc_recei_index[0]],[i*2+1]].values[0][0]
        for i in range(len(acc_recei)):
            RTR[str(self.year - i)+"_RTR"] = round(sales[str(self.year - i)]/acc_recei[str(self.year - i)],3)
        
        return RTR
```

### makeplt function module 

```python
from matplotlib import pyplot as plt
import matplotlib.font_manager as fm 
####    한글폰트설정    ####
font_fname = "C:\\Users\\chon0\\Anaconda3\\envs\\develroom\\lib\\site-packages\\matplotlib\\mpl-data\\fonts\\ttf\\Lotte_Dream_bold.ttf"
fontprop = fm.FontProperties(fname=font_fname)
####    ###########    #### 

def makeplt (diction,title):
    y=[]
    year =[]
    sorted_key = sorted(diction)
    for i in sorted_key:
        #x.append(i)
        y.append(diction[i])
    
    tmp =''
    for i in sorted_key:
        for j in i :
            if str(0)<= j <= str(9):
                tmp = tmp + j
        year.append(int(tmp))
        tmp=''
    
    differ = max(y) - min(y)
    if differ>=100:
        differ = int(round(differ,-1))
        max_lim = round(max(y) + differ*1.5,-2)
        min_lim  = round(min(y) - differ*1.5,-2)
    elif 10<=differ<100:
        differ = int(round(differ,0))
        max_lim = round(max(y) + differ*1.5,-1)
        min_lim  = round(min(y) - differ*1.5,-1)
    elif -10<=differ<10:
        differ = round(differ,1)
        max_lim = round(max(y) + differ*2,1)
        min_lim  = round(min(y) - differ*2,1)


    plt.xticks(year)
    plt.ylim([min_lim,max_lim])   
    
    plt.grid(True)

    # Show the major grid lines with dark grey lines
    #plt.grid(b=True, which='major', color='#999999', linestyle='-')

    # Show the minor grid lines with very faint and almost transparent grey lines
    #plt.minorticks_on()
    #plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=)

    for i in range(len(year)):
        if y[i]>=0:
            plt.text(year[i],y[i],"{}".format(y[i]),fontsize=10)
        else : 
            plt.text(year[i],y[i],"{}".format(y[i]),color='red',fontsize=10)

    plt.title(title, fontproperties=fontprop, size = 18)
    plt.xlabel('(단위 : 년도)',fontproperties=fontprop,position=(1,0))
    plt.ylabel('(단위 : %)',fontproperties=fontprop,position=(0,1))

    plt.plot(year,y)
    #plt.show()
    #print(type(plt.plot(year,y)))
    return plt.plot(year,y) 
```
